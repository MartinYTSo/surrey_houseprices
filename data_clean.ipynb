```python
import pandas as pd
import os
import statistics
import seaborn as sns
from matplotlib import pyplot
from sklearn import *
from  sklearn.model_selection import *
import numpy as np
```

```python
#load and view csv
df=pd.read_csv('data/zillow_surrey.csv')
df.dropna()
# df=df['list-card-price'].astype(str)
df['price']=df['list-card-price'].str[2:]
df['price'] = df['price'].str.replace(r'\D', '')
df = df[df['price'].notna()]
df
```

```python
#check types 
df.dtypes
```

```python
# convert objects to numerical figures
price_int=df['price'].astype(int)
df['priceint']=price_int

df_checktype=df[['priceint', 'list-card-details 3']].rename(columns={'list-card-details 3': 'sqft'}).dtypes
df_clean=df[['priceint', 'list-card-details 3', 'list-card-details', 'list-card-details 2']].rename(columns={'list-card-details 3': 'sqft',
                                                                                   'list-card-details': 'bds', 'list-card-details 2': 'ba',
                                                                                    'priceint':'price'})
df_clean['bds'].replace('Studio',0.5).astype(float)
df_clean['ba'].replace('--',0).astype(float)
df_clean['sqft'].replace(',','', regex=True, inplace=True)

df_checktype
```

```python
# replace NA and misc entries with values
df_clean=df_clean.replace(to_replace ="Studio",
                 value ="0.5").replace(to_replace= "--",value="0").astype(float)
df_clean
```

```python
#write df to csv

df_clean.to_csv('data/surrey_zillow_clean.csv', index=False)
```
